```python
samples = ['S0']*4+['S22']*4+['S23']*4+['S26']*4
num_cells = [536]*4+[404]*4+[211]*4+[798]*4
num_genes = [10,100,500,1000]*4
max_mems = []

for S in ['S0','S22','S23','S26']:
    for n_genes in [10,100,500,1000]:
        raw_mem = pd.read_csv(f'results/memProf/memProf_lu_T_{n_genes}genes_{S}.txt',skiprows=5,sep='\t',header=None)
        samp_max_mem = []
        for r in range(len(raw_mem)):
            ind_ =raw_mem.loc[r].values[0].find('MiB')
            if ind_ != -1:
                val = raw_mem.loc[r].values[0][ind_-7:ind_].replace(" ","")
                samp_max_mem.append(np.float(val))
        max_ = np.max(np.array(samp_max_mem))

        max_mems.append(max_)
```

```python
for i in range(len(max_mems)):

    print(samples[i]+' & '+str(num_cells[i])+' & '+str(num_genes[i])+' & '+str(max_mems[i])+' \\'+'\\')
    print('\hline')
```
